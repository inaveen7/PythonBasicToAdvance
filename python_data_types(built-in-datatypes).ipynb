```python
'''
-> In programming data types is an important concept.

-> Variable can store data of different types and different types can do different things.

-> Python has the following built-in(7) data types by default.

Text type            :  str

Numeric Types        :  int, float, complex

Sequence Types       :  list, tuple, range

Mapping Type         :  dict --> dictory

Set Types            :  set, frozenset

boolean Type         :  bool

binary Types         :  bytes, bytearray, memoryview


'''
```

```python
# Getting Data type

'''
You can get the data type of an object by using type() function.
'''

#Example to print the type of the variable

x=5                # yet we don't know the data type of the variable 'x'

print(type(x))     # here, we are using type function to get the data type of the variable 'x'
```

```python
# Settig the data type

'''
In python, the data type is set when you assign the value to a variable
'''

# Example



x='Hello'                                   # string str type
print(type(x))
print(x+"\n")
 
x=20
print(type(x))                              # intger, int type 
print(x)
print()

x=20.5
print(type(x))                              # float type 
print(x)
print()

x=1j
print(type(x))                              # complect type 
print(x)
print()


x=['Mango', 'Apple', 'Banana']              # list type 
print(type(x))
print(x)
print()

x=('Mango', 'Apple', 'Banana')              # tuple type 
print(type(x))
print(x)
print()

x = range(6)                                #range type 
print(type(x))
print(x)
print()

x={"name" : "Steve", "age" : 120}           #dictonary, dict type 
print(type(x))
print(x)
print()

x= {'Mango', 'Apple', 'Banana'}             # set type
print(type(x))
print(x)
print()

x=frozenset({'Mango', 'Apple', 'Banana'})   # frozen type
print(type(x))
print(x)
print()

x=True                                      # boolen, bool type 
print(type(x))                               
print(x)
print()

x= b"Hello"                                 # binary type, bytes type 
print(type(x))
print(x)
print()

x=bytearray(5)                              # binary type, bytearray type 
print(type(x))
print(x)
print()

x=memoryview(bytes(5))                      # binary type, memoryview type
print(type(x))
print(x)

```

```python
# Setting Specific Data Type

'''
If you want to specify the data type, you can use the following constructor functions.

'''

#Example


x = str("Hello!!, There")                         # string str type
print(x)
print(type(x))
print()


                                                  # intger, int type
x = int(10)
print(x)
print(type(x))
print()



x = float(20.5)                                   # float type 
print(x)
print(type(x))
print()



x = complex(5j)                                   # complect type         
print(x)
print(type(x))
print()



x = list(('Mango', 'Apple', 'Banana'))            # list type
print(x)
print(type(x))
print()



x = tuple (('Mango', 'Apple', 'Banana'))          # tuple type 
print(x)
print(type(x))
print()



x = range(6)                                       #range type 
print(x)
print(type(x))
print()



x = dict(name = 'Steve', age=120)                  #dictonary, dict type      
print(x)
print(type(x))
print()



x = set(('Mango', 'Apple', 'Banana'))              # set type
print(x)
print(type(x))
print()

x = frozenset(('Mango', 'Apple', 'Banana'))        # frozen type 
print(x)
print(type(x))
print()



x = bool(0)                                        # boolen, bool type (False)
print("x = 0 = " , x )
print(type(x))


y = bool(5)                                        # boolen, bool type (True)
print("y = 5 = " , y )
print(type(y))
print()




x = bytes(6)                                        # binary type, bytes type       
print(x)
print(type(x))
print()



x = bytearray(6)                                     # binary type, bytearray type 
print(x)
print(type(x))
print()



x = memoryview(bytes(6))                             # binary type, memoryview type
print(x)
print(type(x))




```

```python
# Python Numbers

'''
-> There are 3 numeric types in python
    
    1. int 
    2. float
    3. complex
    
'''

#Variables of numeric type are created when you assign a value to them 

#Example 

x = 1        # int 

y = 2.7      # float

z = 5j       # complex 


# To verify the type of an object, use the type() function

print(type(x), '\n')

print(type(y), '\n')

print(type(z), '\n')

```

```python
# Int : int, or integer is a whole number, positive or negative, with out any decimals, of unlimited length

x = 1

y = 35834738

z = -7343

print('x = ',x, type(x), '\n')

print("y = ",y, type(y), '\n')

print('z = ',z, type(z), '\n')

```

```python
# Float : float, or floating pointer number is a number, positive or negative, containing one or more decimals

x = 1.0

y = 1.10

z = -35.97


print('x = ',x, type(x), ' \n')

print("y = ",y, type(y), ' \n')

print('z = ',z, type(z), ' \n')
```

```python
# Complex : complex number is a number with a 'j' as the imaginary part.

x = 3+5j

y = 7j          # here, [ 7j = 7+0j ]

z = -5j         # here, [ -5j = -5+0j ]

print('x = ',x, type(x), ' \n')

print("y = ",y, type(y), ' \n')

print('z = ',z, type(z), ' \n')
 

a = 7              # int
    
n= complex(a)      #a=7

print("priting a=7j in actual complex format", n)
```

```python
# Type Conversion

'''
-> We can convert one type to an other type by using ' int(), float(), complex()' methods

'''

# converting from one type to an other

x = 1

y = 2.8 

z = 3j

print(x,'\n',y,'\n',z,'\n')


a = float(x)               # coverting from int to float
print(a, type(a), ' \n')


b = int(y)                # conveting from float to int
print(b, type(b), ' \n')


c = complex(x)        
print(c, type(c), ' \n')  # converting from int to complex 


d = complex(y)
print(d, type(d), ' \n')  # coverting from float to complex


'''
 NOTE :- We can not convert  Complex number into an other number type
'''
```
